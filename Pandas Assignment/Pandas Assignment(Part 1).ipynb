```python
import pandas as pd
```

```python
import os
```

```python
import glob
```

```python
import numpy as np
```

```python
  
all_files = glob.glob(os.path.join("*.csv"))     

df_from_each_file = (pd.read_csv(f) for f in all_files)
all_states   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
us_census = pd.DataFrame(all_states)
```

```python
us_census
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census.rename( columns={'Unnamed: 0':'Serial'})
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census.drop(columns = ['Unnamed: 0'],inplace= True)
```

```python
us_census.head()
```

```python
us_census[['MalePop', 'FemalePop']]= us_census.GenderPop.str.split('_',expand=True)
```

```python
us_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
us_census.head()
```

```python
us_census['Income'] = us_census.Income.str.strip('$')
```

```python
us_census = us_census.replace('%', '', regex = True)
```

```python
us_census.head(20)
```

```python
us_census['FemalePop'] = us_census.FemalePop.replace('F','',regex=True)
```

```python
us_census['MalePop'] = us_census.MalePop.replace('M','',regex=True)
```

```python
us_census.dtypes
```

```python
us_census.loc[:,'Hispanic':'Income']= round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
us_census.head()
```

```python
us_census['TotalPop']= us_census['TotalPop'].astype(int)
```

```python
us_census['MalePop']= us_census['MalePop'].astype(int)
```

```python
# us_census['FemalePop'] = us_census.FemalePop.replace(r'^\s*$', np.nan, regex=True)
```

```python
# us_census.FemalePop.isnull()
```

```python
# us_census['FemalePop']= us_census.FemalePop.fillna(lambda x : x (us_census['TotalPop']-us_census['MalePop']))
```

```python
us_census.drop(columns = ['FemalePop'],inplace= True)
```

```python
us_census['FemalePop'] = us_census['TotalPop'] - us_census['MalePop']
```

```python
us_census.drop_duplicates(inplace = True)
```
